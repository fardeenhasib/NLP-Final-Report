```python
label_file='labels.txt'
label_data=open(label_file,'r').readlines()
for i in range(len(label_data)):
  label_data[i]=label_data[i].replace('\n','')
  if label_data[i]=='entailment':
    label_data[i]=1
  else:
    label_data[i]=0
```

```python
print(label_data)
```

```python
data_final=[]
count=0
data_file='processed_text.txt'
data=open(data_file,'r').readlines()
hypothesises,premises=[],[]
for i in range(0,len(data),2):
  hypothesis1=data[i].split('\t')[0]
  premise1=data[i].split('\t')[1]

  hypothesis2=data[i+1].split('\t')[0]
  premise2=data[i+1].split('\t')[1]
  if premise1==premise2:
   
    if label_data[i]==1 and label_data[i+1]==0:
        data_final.append([premise1,hypothesis1,hypothesis2])
        count+=1
    elif label_data[i]==0 and label_data[i+1]==1:
        data_final.append([premise1,hypothesis2,hypothesis1])
        count+=1
    
  

print(count*2)
print(len(data))
```

```python
import pandas as pd
data_final=pd.DataFrame(data=data_final,columns=['sent0','sent1','hard_neg'])
data_final.to_csv('our_data_with_hard_negatives.csv')
```
